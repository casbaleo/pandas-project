# Data Cleaning in pandas

```python
df = pd.read_excel(r"C:\Users\Leo\Desktop\pandas-project\Customer Call List.xlsx")
df
```

```python
#Workflow. 1. Eliminate duplicates
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = 'Not_Useful_Column') ##Workflow. 2. Eliminate unnecessary columns --Last Name
df
```

```python
#worflow 3. Clean strings opcion 1
#df['Last_Name'].str.strip() #strip es equivalente a LEFT/RIGHT en SQL - lstrip, rstrip, si hago strip() quita espacios en blanco, pero puedo espeficiar los caracteres a eliminar
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')
#df
```

```python
#worflow 3. Clean strings opcion 2 --Last Name
df['Last_Name'] = df['Last_Name'].str.strip('./_') #puedo eliminar todos los caracteres en un solo comando
df
```

```python
#worflow 4. Clean strings --Phone number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '') #Reemplazar todo lo que este entre a-z, A-Z y 0-9 por ''
df
```

```python
#worflow 4. Clean strings --Phone number
#df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x)) #primero debemos transformar cada valor en string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] )
df
```

```python
#worflow 4. Clean strings --Phone number
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '') #eliminar los nan--
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '') #eliminar los Na--
df
```

```python
#worflow 5. Split Address in 3 columns
df[['Street_Adress', 'State', 'ZIP']] = df['Address'].str.split(',', 2, expand = True)
df
```

```python
#worflow 6. Clean Paying Customer/Do not Contact
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
#Elimino columna address
df = df.drop(columns='Address')
```

```python
#worflow 7. Clean n/as
df = df.replace('n/a', '')
df = df.replace('N/a', '')
df = df.fillna('')
df
```

```python
#worflow 8. Remove rows which don't have phone number (do not contact = Y)
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)

df

#PARA cada x dentro del indice, si el indice en la columna DonotContact es Y, entonces lo borras y me aplicas los cambios
```

```python
#worflow 8. Remove rows which don't have phone number (there's no phone number)
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
        
df

#another way to drop values
#df = df.dropna(subset = 'Phone_Number'), inplace true)
```

```python
#worflow 8. Reset index
df = df.reset_index(drop = True) 
df
```
